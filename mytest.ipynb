## Define parameters

```python

FOVumLat = 66.6
FOVpxLat = 512.0 # 512 
pxPerUmLat = FOVpxLat/FOVumLat 
pxPerUmAx = 0.2 # 2.0 
wavelength = 915.0 
NA = 0.95 
windowUm = [11, 4, 4]
options = {'FOVumLat':FOVumLat, 'FOVpxLat':FOVpxLat, 'pxPerUmLat':FOVpxLat/FOVumLat, 'pxPerUmAx':pxPerUmAx, 'wavelength':970.0, 'NA':0.6, 'windowUm':windowUm} 
options['thresh'] = .2
```

## Get PSF

```python
im = imread('J:/Vasily/20200126/2ppsf_915nm/3_915nm_corrected_d=0.2um_zoom15_power10mW(thorlabs)_-1987--1997um.tif', plugin='tifffile')
print(im.shape)
for i in range(0, im.shape[0]):
       im[i] = im[i] - np.min(im[i])
```

```python
 beads, maxima, centers, smoothed = getCenters(im, options)
```

```python
plt.figure(figsize=(10,10));
plt.imshow(smoothed);

plt.plot(centers[:, 2], centers[:, 1], 'w.', ms=3);
plt.xlim([0, smoothed.shape[0]])
plt.ylim([smoothed.shape[1], 0])
plt.axis('off');
```

```python
data, beads, maxima, centers, smoothed = compute(im, options)
```

```python
PSF = pd.concat([x[0] for x in data])
PSF['Max'] = maxima
PSF = PSF.reset_index().drop(['index'],axis=1)
latProfile = [x[1] for x in data]
axProfile = [x[2] for x in data]
```
